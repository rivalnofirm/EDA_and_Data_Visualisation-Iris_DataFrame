## Iris Dataset : Data Visualisation

#### Import Library

```python
import matplotlib.pyplot as plt # Library matplotlib digunakan untuk melakukan visualisasi data
import seaborn as sns # Library seaborn digunakan untuk melakukan visualisasi data
```

#### Bar Plot

```python
df_iris['Species'].value_counts() # digunakan untuk menghitung jumlah dari spesies yang ada
```

```python
# Bar Plot menggunakan library Matplotlib
df_iris['Species'].value_counts().plot.bar() # pemannggilan .plot().bar() digunakan untuk menampilkan visualisasi menggunakan Bar Plot
plt.tight_layout() # method tight_layout() berfungsi untuk menyesuaikan hasil visualisasi sesuai layar yg tersedia
plt.show() # baris ini berfungsi untuk menampilkan hasil visualisasi
```

```python
# Bar Plot menggunakan library Seaborn
sns.countplot(data=df_iris, x='Species') # digunakan untuk menapilkan bar plot dengan spesifik harus di tuliskan parameter x axis
plt.tight_layout()
plt.show()
```

#### Pie Chart

```python
df_iris['Species'].value_counts().plot.pie(autopct='%1.1f%%', labels=None, legend=True)
plt.tight_layout()
plt.show()
```

#### Line Chart

```python
df_iris.plot()
plt.tight_layout()
plt.show()
```

#### Histogram

```python
df_iris.hist(figsize=(6,6), bins=10)
plt.tight_layout()
plt.show()
```

#### Boxplot

```python
df_iris.boxplot()
plt.tight_layout()
plt.show()
```

#### Scatter Plot

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=df_iris, hue='Species')
plt.tight_layout()
plt.show()
```

#### Heatmap

```python
sns.heatmap(data=df_iris.corr())
plt.tight_layout()
plt.show()
```

#### Pair Plot

```python
sns.pairplot(df_iris, hue='Species', markers='+')
plt.tight_layout()
plt.show()
```

#### Violin Plot

```python
sns.violinplot(data=df_iris, y='Species', x='SepalLengthCm', inner='quartile')
plt.tight_layout()
plt.show()
```
